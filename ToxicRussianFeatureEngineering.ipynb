```
from string import punctuation

import pandas as pd

from nltk.corpus import stopwords

import stanza
from spacy_stanza import StanzaLanguage
```

```
#stanza.download('ru')
snlp = stanza.Pipeline(lang="ru")
nlp = StanzaLanguage(snlp)
```

```
toxic_rus = pd.read_pickle("toxic_main.pickle")
toxic_rus.head()
```

```
my_toxic = pd.DataFrame(toxic_rus["comment"].iloc[0:2000])
my_toxic.head()
```

```
my_toxic.shape
```

```
def is_noun(text):
    doc = nlp(text)    
    NOUNS = []
    for token in doc:
        if token.pos_ == "NOUN":
            NOUNS.append(str(token).lower())
    NOUNS = set(NOUNS)
    #print(doc,"NOUNS:", NOUNS)
    return " ".join(NOUNS)
```

```
my_toxic["nouns"] = my_toxic["comment"].apply(is_noun)
```
